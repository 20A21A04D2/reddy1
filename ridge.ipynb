```python
data1=data.drop(['lat','ID','lon'],axis=1)#remove column of lat,id,lon by using drop function
```

```python
data1
```

```python
data2=pd.get_dummies(data1)#where the lounge model it shows"1" other models it shows "0"
data2
```

```python
data2.shape#how many rows and columns in the data frame
```

```python
z=data2.loc[(data.model=='lounge')]#determine only for lounge cars
```

```python
z
```

```python
y=z['price']#removing price from data2 and put in new data frame y
x=z.drop(['price'],axis=1)#remaining data can be put in another data fram
```

```python
x# to get data in x
```

```python
y# to get data in y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)#dividing training data and testing data
```

```python
x_test.head(5)#display top 5 data in testing data
```

```python
y_test.head(5)#display top 5 data in testing data price dataframe
```

```python
x_train.head(5)#display top 5 data in training data
```

```python
y_train.head(5)#display top 5 data in training data price dataframe
```

```python
from sklearn.model_selection import GridSearchCV#ridge  regression
from sklearn.linear_model import Ridge

alpha = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20,30]

ridge = Ridge()

parameters = {'alpha': alpha}

ridge_regressor = GridSearchCV(ridge, parameters)

ridge_regressor.fit(x_train, y_train)
```

```python
ridge_regressor.best_params_#alpha value
```

```python
ridge=Ridge(alpha=30)
ridge.fit(x_train,y_train)
y_pred_ridge=ridge.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error#mean_squared error
Ridge_Error=mean_squared_error(y_pred_ridge,y_test)
Ridge_Error
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_ridge)#finding the efficieny
```

```python
Results=pd.DataFrame(columns=['Actual','predicted'])##creating a data frame called results for given price and predicted price
Results['Actual']=y_test
Results['predicted']=y_pred_ridge
Results=Results.reset_index()
Results['ID']=Results.index
Results.head(10)
```

```python
import seaborn as sns#putting the graoh for actual price and predicted price
import matplotlib.pyplot as plt
sns.lineplot(x='ID',y='Actual',data=Results.head(50))
sns.lineplot(x='ID',y='predicted',data=Results.head(50))
plt.plot()
```
